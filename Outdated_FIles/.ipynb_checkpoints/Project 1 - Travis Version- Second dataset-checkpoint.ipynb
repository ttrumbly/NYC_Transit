def get_data(week_nums):
    url = ""
    dfs = []
    for week_num in week_nums:
        file_url = url.format(week_num)
        dfs.append(pd.read_csv(file_url))
    return pd.concat(dfs)
        
week_nums = [180303]
turnstiles_df = get_data(week_nums)

```python
filename = 'turnstile_180303.csv'
subwaydf = pd.read_csv(filename)
```

```python
subwaydf.columns = subwaydf.columns.str.strip()
```

```python
subwaydf['datetime'] = subwaydf[['DATE','TIME']].apply(lambda x: ' '.join(x), axis=1)
subwaydf['datetime'] = pd.to_datetime(subwaydf['datetime'])
```

```python
subwaydf['4HR_Entries'] = subwaydf['ENTRIES'] - subwaydf['ENTRIES'].shift(1)
subwaydf['4HR_Exits'] = subwaydf['EXITS'] - subwaydf['EXITS'].shift(1)
```

```python
subwaydf['4HR_Entries'] = subwaydf['4HR_Entries'].where(subwaydf['SCP'] == subwaydf['SCP'].shift(1))
subwaydf['4HR_Exits'] = subwaydf['4HR_Exits'].where(subwaydf['SCP'] == subwaydf['SCP'].shift(1))
```
