```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
filename = "cleaned_4hr_entries.csv"
month = pd.read_csv(filename)
```

```python
filename = "top100_station.csv"
top_station = pd.read_csv(filename)
```

```python
top_station = top_station.head(11)
```

```python
month['datetime'] = pd.to_datetime(month['datetime'])
month['weekday'] = month['datetime'].dt.weekday
month['weekday2'] = month['datetime'].dt.weekday_name
```

```python
month['date'] = month['datetime'].dt.date
month['time'] = month['datetime'].dt.time
```

```python
month1 = month.loc[(month['weekday2'] == 'Monday') & month['station'].isin(top_station['station'])]
```

```python
month2 = month1.groupby(['station', 'linename', 'time'], as_index=False)['hr4'].sum()
month2.set_index('time', inplace=True)
```

```python
month2.groupby(['station', 'linename']).plot(legend=True)
```
